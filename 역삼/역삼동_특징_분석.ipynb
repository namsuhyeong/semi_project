```python
# 행정동별 DF 만드는 함수
def dong_df(df, dong):
    temp_dong = df.loc[df.행정동.isin([dong]), :]
    return temp_dong
```

```python
# 상권코드명 DF
def 상권코드_df(df, code):
    temp = df.loc[df.상권_코드_명.isin([code]), :]
    return temp
```

```python
# 기준년분기로 추출하는 함수
def 년분기_df(df, year):
    temp = df.loc[df.기준_년분기_코드.isin([year]), :]
    return temp
```

```python
# 상권구분코드별 상권_코드_명
def code_df(df, code):
    temp = df.loc[df.상권_구분_코드_명.isin([code]), ['상권_구분_코드_명', '상권_코드_명']]
    code_list = temp.상권_코드_명.unique().tolist()
    return code_list
```

```python
# 상권코드명 추출하는 함수
def 상권코드명(df):
    상권_list = df.상권_코드_명.unique().tolist()
    return 상권_list
```

```python
# 행정동 직장인구 DF를 상권코드명별 분기별 인구 DF로 만드는 함수
def 상권분기별인구(df, col):
    group = df.groupby(['상권_코드_명', '기준_년분기_코드'])[col].sum()
    return group
```

```python
# 분기별 업종별 DF
def 분기별업종별(df, col):
    temp = df.groupby(['기준_년분기_코드', '서비스_업종_코드_명'])[col].sum()
    return temp
```

```python
# 년분기별로 group
def 년분기(df, col):
    temp = df.groupby('기준_년분기_코드')[col].sum()
    return temp
```

```python
# 이상치 확인 함수
def find_outlier(df):
    print('이상치 확인하기\n')
    
    # 이상치 판별하지 않을 컬럼 제외하기
    no_finding = ['행정동', '행정동_코드_명', '기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '서비스_업종_코드_명', '서비스_업종_코드']

    # DF의 컬럼들을 돌아가며 이상치 분석하기
    for column in df.columns:
        if column in no_finding:
            continue
            
        # 1단계 : 사분위수 가져오기
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # 임계값 가져오기
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # 이상치 구분하기
        # 이상치인 조건
        condition = (df[column] < lower_bound) | (df[column] > upper_bound)

        # 만약 이상치가 없다면? (= 이상치 조건이 모두 False)
        if condition.sum() == 0:
            print(f'{column} : 이상치 없음')
        outlier = df.loc[condition, :].reset_index(drop=True)
            
        return outlier
```

```python
# 이상치 변경 함수
def change_outlier(df):
    print('이상치 확인하기\n')
    
    # 이상치 판별하지 않을 컬럼 제외하기
    no_finding = ['행정동', '행정동_코드_명', '기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '서비스_업종_코드_명', '서비스_업종_코드']

    # DF의 컬럼들을 돌아가며 이상치 분석하기
    for column in df.columns:
        if column in no_finding:
            continue
            
        # 1단계 : 사분위수 가져오기
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # 임계값 가져오기
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # 이상치 구분하기
        condition = (df[column] < lower_bound) | (df[column] > upper_bound)
        
        # 만약 이상치가 없다면? (= 이상치 조건이 모두 False)
        if condition.sum() == 0:
            print(f'{column} : 이상치 없음')
        else:
            print(f'{column} : 이상치 {condition.sum()}개')
        outlier = df.loc[condition, :].reset_index(drop=True)
        
        # 이상치를 임계값에 넣어주기
        df[column] = df[column].apply(lambda x : min(x,  upper_bound))
        df[column] = df[column].apply(lambda x : max(x,  lower_bound))
        
        # 변경된 값으로 된 DF를 새로운 변수에 저장하기
        change_df = df.reset_index(drop=True)
    
    return change_df
```
